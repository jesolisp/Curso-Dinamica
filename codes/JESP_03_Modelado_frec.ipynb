## Función de transferencia

Considere un sistema lineal invariante en el tiempo (LTI) cuya función de transferencia es $G(s)$ y la entrada así como la salida se representan por $x(t)$, $y(t)$ respectivamente. Dicha función se puede escribir como el cociente de dos polinomios en $s$, esto es

\begin{equation}
 G(s) = \frac{p(s)}{q(s)} = \frac{p(s)}{(s+s_{1})(s+s_{2})\cdots(s+s_{n})},
\end{equation}

o bien

\begin{equation}
 G(s) = \frac{b_{n}s^{n} + \cdots + b_{3}s^{3} + b_{2}s^{2} + b_{1}s + b_{0}}{a_{n}s^{n} + \cdots + a_{3}s^{3} + a_{2}s^{2} + a_{1}s + a_{0}}.
\end{equation}

La transformada de Laplace de la salida $Y(s)$ es

\begin{equation}
 Y(s) = G(s)X(s) = \frac{p(s)}{q(s)}X(s),
\end{equation}

donde $X(s)$ es la transformada de Laplace de la entrada $x(t)$.

La respuesta en estado estacionario de un sistema estable LTI a una entrada sinusoidal no depende de las condiciones iniciales. Considerando que $Y(s)$ tiene  únicamente polos distintos (simples), la ecuación anterior se puede representar en fracciones parciales como sigue:

\begin{equation}
 Y(s) = G(s)X(s) = G(s) \frac{\omega X}{s^{2} + \omega^{2}},
\end{equation}

\begin{equation}
 Y(s) = \frac{a}{s + j\omega} + \frac{\bar{a}}{s - j\omega} + \frac{b_{1}}{s + s_{1}} + \frac{b_{2}}{s + s_{2}} + \dots + \frac{b_{n}}{s + s_{n}},
\end{equation}

donde $a$ y $b_{i},i=1,2,\dots,n$ son constantes y $\bar{a}$ es el complejo conjugado de $a$.

Por tanto, aplicando la transformada inversa de Laplace de la ecuación tenemos

\begin{equation}
 y(t) = ae^{-j\omega t} + \bar{a}e^{j\omega t} + b_{1}e^{-s_{1} t} + b_{2}e^{-s_{2} t} + \cdots + b_{n}e^{-s_{n} t}, \quad t\geq 0.
\end{equation}

### Función de transferencia propia

Considere la siguiente función de transferencia

\begin{equation}
 G(s) = \frac{N(s)}{D(s)},
\end{equation}

donde $N(s)$ y $D(s)$ son dos polinomios con coeficientes reales, entonces tenemos
* Función **impropia** si $\text{deg }N(s) > \text{deg }D(s)$. 
* Función **propia** si $\text{deg }N(s) \leq \text{deg }D(s)$.
* Función **estrictamente propia** si $\text{deg }N(s) < \text{deg }D(s)$.
* Función **bipropia** si $\text{deg }N(s) = \text{deg }D(s)$.

### Estabilidad de sistemas

* **Sistema Absolutamente Estable.** todos los polos del sistema estén estrictamente en el semi-plano izquierdo, sin importar la multiplicidad.
* **Sistema Marginalmente Estable.** al menos un polo está sobre el eje imaginario con multiplicidad uno y el resto están estrictamente en el semi-plano izquierdo, sin importar la multiplicidad.
* **Sistema Inestable.** al menos un polo está sobre el eje imaginario con multiplicidad mayor a uno o en el semiplano derecho.

### Polinomio de Hurwitz

**Definición 2.** Un polinomio con coeficientes reales es llamado polinomio de **Hurwitz** si todas sus raíces tienen partes reales negativas. 

Considere el siguiente polinomio

\begin{equation}
 D(s) = a_{n}s^{n} + a_{n-1}s^{n-1} + \dots + a_{1}s + a_{0}, \quad a_{n}>0,
\end{equation}

donde $a_{i},i=0,1,\dots,n$ son constantes reales.

#### Condiciones necesarias

* Si $D(s)$ es Hurwitz, entonces todos los coeficientes de $D(s)$ deben ser positivos.
* Si $D(s)$ tiene un término perdido o un coeficiente cero, entonces no es Hurwitz.

**Teorema 1.** Un polinomio con coeficientes positivos es un polinomio Hurwitz si y sólo si cada entrada en la tabla Routh es positivo, o equivalentemente, si y sólo si cada entrada en la primer columna de la tabla $(b_{61},b_{51},b_{41},b_{31},b_{21},b_{11},b_{01})$ es positivo.

### Movimiento de traslación

**Masa.** Propiedad de un elemento de almacenar energía cinética del movimiento de traslación (Kuo, 1996).

Ley de movimiento de Newton. La suma de todas las fuerzas que actúan sobre un cuerpo rígido dada una dirección es igual al producto de la masa del cuerpo en cuestión por su aceleración en la misma dirección (Kuo, 1996)

\begin{equation}
 \sum F = m\cdot a.
\end{equation}

La ecuación de la fuerza está dada como sigue

\begin{equation}
 \quad f(t) = M\cdot a(t).
\end{equation}

Si denotamos a $y(t)$ como la posición de una partícula a lo largo del tiempo, tenemos

\begin{equation}
 f(t) = M\cdot \frac{\mathrm{d}^{2}y(t)}{\mathrm{d}t^{2}} \equiv M \cdot \frac{\mathrm{d}v(t)}{\mathrm{d}t},
\end{equation}

donde $v(t)$ denota la velocidad.

**Resorte lineal.** Elemento que almacena energía potencial.
Si la deformación es pequeña, podemos aproximar su comportamiento por la siguiente relación

\begin{equation}
 f(t) = Ky(t),
\end{equation}

donde $K$ es la *constante del resorte* (rigidez).

Si el resorte es precargado con una tensión $T$, tenemos

\begin{equation}
 f(t) - T = Ky(t).
\end{equation}


Las fuerzas de fricción se encuentran en sistemas  físicos cuando existe movimiento entre dos elementos físicos. En su mayoría son de naturaleza no lineal y tenemos tres tipos:

1. **Fricción viscosa.** Representa una fuerza como una relación lineal entre la fuerza aplicada y la velocidad. Este tipo se representa como un amortiguador y su expresión es la siguiente

\begin{equation}
 f(t) = B \frac{\mathrm{d}y(t)}{\mathrm{d}t} \equiv B \dot{y}(t),
\end{equation}

donde $B$ es el coeficiente de fricción viscosa con unidades $N/m/s$.

2. **Fricción estática.** Representa una fuerza que tiende a prevenir el movimiento desde el comienzo. Su ecuación es

\begin{equation}
 f(t) = \pm \left. (F_{s}) \right|_{\dot{y} = 0}.
\end{equation}


3. **Fricción de Coulomb.** Es una fuerza que tiene una amplitud constante con respecto al cambio de velocidad. La ecuación que la representa está dada como sigue

\begin{equation}
 f(t) = F_{c} \frac{\dot{y}(t)}{|\dot{y}(t)|},
\end{equation}

donde $F_{c}$ es el coeficiente de la fricción de Coulomb.

### Movimiento de rotación

Se define como el movimiento alrededor de un eje fijo. Nos dice que está la suma de los momentos o pares alrededor de un eje fijo es igual al producto de la inercia por la aceleración angular

\begin{equation}
 \sum F = J\cdot \alpha,
\end{equation}

donde $J$ denota la inercia, $a$ la aceleración angular. Otras variables utilizadas para el movimento rotacional son: par ($\tau$), velocidad angular ($\omega$) y desplazamiento angular ($\theta$).

**Inercia.** Propiedad de un elemento de almacenar energía cinética del movimento de rotación. Depende de la composición geométrica al rededor del eje de rotación. 

La ecucación que describe el par aplicado a un cuerpo con inercia $J$, está dada como sigue

\begin{equation}
 \tau(t) = J\cdot \alpha(t) \equiv J \dot{w}(t) \equiv J \ddot{\theta}(t),
\end{equation}

donde $\theta(t)$ es el desplazamiento angular, $\omega(t)$ la velocidad angular y $\alpha(t)$ la aceleración angular.

Retomando la segunda Ley de Newton para el caso de la masa

\begin{equation}
 F_{m} = m\cdot a_{m}.
\end{equation}

Retomando la Ley de Hooke

\begin{equation}
 F_{k} = k\cdot x.
\end{equation}

Tenemos

\begin{equation}
 \sum F_{x} = 0.
\end{equation}

Considerando las dos fuerzas, tenemos

\begin{equation}
 m\cdot a_{m} + k\cdot{x} = 0,
\end{equation}

donde $a_{m}$ puede ser representada como

\begin{equation}
 a_{m} = \frac{\mathrm{d}^{2}x}{\mathrm{d}t^{2}} \equiv \ddot{x},
\end{equation}

sustituyendo tenemos

\begin{equation}
 m\cdot \ddot{x} + k\cdot x = 0,
\end{equation}

donde $m$ es la masa y $k$ la constante del resorte (rigidez).

Para obtener una representación en función de transferencia, aplicamos la transformada de Laplace *i.e.*

\begin{equation}
 \mathcal{L} \left\{ m \cdot\ddot{x} + k \cdot x \right\} = 0,
\end{equation}

teniendo como resultado la siguiente representación

\begin{equation}
 X(s) = \frac{m\cdot x_{0}\cdot s}{m\cdot s^{2} + k} = \frac{x_{0}\cdot s}{s^{2} + \frac{k}{m}}.
\end{equation}

Aplicando la transformada inversa de Laplace, obtenemos

\begin{equation}
 \mathcal{L}^{-1} \left\{ \frac{x_{0}\cdot s}{s^{2} + \frac{k}{m}} \right\} = x_{0}\cdot \cos\left( \sqrt{\frac{k}{m}}t \right),
\end{equation}

que corresponde a la ecuación de la posición

\begin{equation}
 x(t) = x_{0}\cdot \cos\left( \sqrt{\frac{k}{m}}t \right),
\end{equation}

derivando la ecuación anterior, obtenemos la ecuación de la velocidad *i.e.*

\begin{equation}
 \dot{x}(t) = -x_{0}\sqrt{\frac{k}{m}}\sin\left(\sqrt{\frac{k}{m}}t \right).
\end{equation}

### Enfoque Euler-Lagrange

**Analogía de movilidad**
* velocidad es análogo a esfuerzo
* fuerza es análogo a flujo

#### Masa traslacional

Una masa traslacional pura es un objeto rígido el cual se mueve a través de un entorno no disipativo. De acuerdo con la segunda Ley de Newton, el momento $p$ de la masa está linealmente relacionado a la velocidad del objeto como se da a continuación

\begin{equation}
 p = mv,
\end{equation}

donde $m$ es la masa Newtoniana del objeto. El momento $p$, está definido por

\begin{equation}
 p = \int_{t_{0}}^{t} F\mathrm{d}t + p(t_{0}), \quad F = \frac{\mathrm{d}p}{\mathrm{d}t},
\end{equation}

donde $F$ denota fuerza.

De acuerdo con la analogía de movilidad, la cantidad de momento es formalmente análoga a la acumulación de flujo. Por lo tanto, una masa pura translacional puede ser clasificada como un almacenamiento de flujo.

La masa Newtoniana, tiene una relación constitutiva lineal intríseca, por consiguiente, la energía almacenada $U$ (**energía cinética**) es igual

\begin{equation}
 U = \frac{1}{2}mv^{2}.
\end{equation}

#### Resorte traslacional
Un resorte traslacional es un objeto mecánico que cuando está sujeto a una fuerza se comprime o alarga sin aceleración significativa de sus componentes, o pérdida de energía debido a la fricción o falta de la deformación. El mecanismo de almacenamiento de energía de un resorte puro es el desplazamiento neto del resorte desde su estado de reposo. Por lo tanto, la variable desplazamiento se define como

\begin{equation}
 x ==  \int_{t_{0}}^{t}v\mathrm{d}t + x(t_{0}), \quad v = \frac{\mathrm{d}x}{\mathrm{d}t}.
\end{equation}

Un resorte lineal ideal obedece la ley de Hooke y tiene una relación constitutiva

\begin{equation} 
 x = \frac{1}{k}F,
\end{equation}

donde $k$ se conoce como la rigidez del resorte.

De acuerdo con la analogía de movilidad, un resorte es un almacén de esfuerzo, ya que el desplazamiento  representa una acumulación de la variable esfuerzo (velocidad). Por lo tanto, la energía almacenada en un resorte es $T$ (**energía potencial**) y puede ser evaluada por la siguiente relación constitutiva

\begin{equation}
 T = \frac{1}{2k}F^{2}.
\end{equation}

#### Disipación traslacional
Un disipador puro es aquel en el que los fenómenos de almacenamiento de energía cinética y potencial están ausentes. Por lo tanto, un objeto rígido y liviano que se mueve a través de un fluido viscoso o se desliza sobre una superficie rugosa tendrá una relación constitutiva que relaciona estáticamente la fuerza aplicada y la velocidad relativa del objeto.

La potencia absorbida por un disipador es el producto de las variables esfuerzo y caudal, y se obtiene de la relación constitutiva como la suma del contenido y cocontenido del disipador. La relación constitutiva de un disipador lineal está dada como sigue

\begin{equation}
 J = G = \frac{1}{2}bv^{2}.
\end{equation}

#### Ecuaciones de Lagrange
Cualquier sustema debe satisfacer las ecuaciones de Lagrange dadas por

\begin{equation}
 \frac{\mathrm{d}}{\mathrm{d}t}\left[\frac{\partial L}{\partial\dot{q}} \right] - \frac{\partial L}{\partial q} + \frac{\partial J}{\partial \dot{q}} = f_{0},
\end{equation}

donde $L(q,\dot{q})$ está definido como

\begin{equation}
 L(q,\dot{q}) = U^{*}(\dot{q})-T(q).
\end{equation}

Además
* $U^{*}$, representa la coenergía total en las reservas de flujo del sistema expresada como una función de las coordenadas de esfuerzo generalizado.
* $T$, la energía total en los almacenes de esfuerzo del sistema expresada como una función de las coordenadas de acumulación de esfuerzo generalizadas.
* $J$, el co-contenido total en los disipadores del sistema expresado como una función de las coordenadas de esfuerzo generalizado.

Para un sistema mecánico **no conservativo**, las ecuaciones de Lagrange están dadas como sigue

\begin{equation}
 \frac{\mathrm{d}}{\mathrm{d}t}\left[\frac{\partial L}{\partial\dot{q}_{j}} \right] - \frac{\partial L}{\partial q_{j}} + \frac{\partial J}{\partial \dot{q}_{j}} = F_{j}, \quad j = 1,\dots,l.
\end{equation}


Además, $F_{j}$ son las fuerzas generalizadas y $q$ es la coordenada generalizada. Dicha coordenada es la que será utilizada para describir el comportamiento del sistema a partir de esta variable.

**Observación.** En este contexto conservativo significa libre de disipación y entradas de fuerzas externas.

#### Principio D'Alembert's
La energía co-cinética del sistema puede ser escrita como:

\begin{equation}
 U^{*}(\dot{q}_{1},\dot{q}_{2},\dots,\dot{q}_{m}) = \sum_{j=1}^{m}U^{*}_{j}(\dot{q}_{j}).
\end{equation}

#### Metodología
Definimos la coordenada generalizada. En este caso, nos interesa el desplazamiento denotado como $x$, por consiguiente tenemos

\begin{equation}
 q = x, \quad \dot{q} = \frac{\mathrm{d}q}{\mathrm{d}t} \equiv \dot{x}.
\end{equation}

Definimos como fuerza generalizada a $u$, *i.e.*

\begin{equation}
 F_{i} = u.
\end{equation}

La energía co-cinética del sistema está dada por

\begin{equation}
 U^{*} = \frac{1}{2}mv^{2} \equiv \frac{1}{2}m\dot{q}^{2}.
\end{equation}

La energía potencial del sistema por

\begin{equation}
 T = \frac{1}{2}kx^{2} \equiv \frac{1}{2}kq^{2}.
\end{equation}

El Lagrangiano del sistema entonces como

\begin{equation}
 L = U^{*} - T,
\end{equation}

\begin{equation}
 L = \frac{1}{2}m\dot{q}^{2} - \frac{1}{2}kq^{2}.
\end{equation}

Obtenemos las derivadas parciales

\begin{equation}
 \frac{\partial L}{\partial \dot{q}} = m\dot{q},
\end{equation}

\begin{equation}
 \frac{\partial L}{\partial q} = -kq,
\end{equation}

\begin{equation}
 \frac{\mathrm{d}}{\mathrm{d}t}\left[ \frac{\partial L}{\partial \dot{q}} \right]= m\ddot{q},
\end{equation}

\begin{equation}
 \frac{\partial J}{\partial q} = b\dot{q}.
\end{equation}

Sustituyendo, tenemos

\begin{equation}
 m\ddot{q} - (-kq) + b\dot{q} = 0,
\end{equation}

\begin{equation}
 m\ddot{q} + b\dot{q} + kq = 0.
\end{equation}

## Modelado de sistemas eléctricos y electro-mecánicos

La relación que establece el flujo electromagnético $\phi$ y la corriente $i$ que lo produce está dada por la siguiente ecuación

\begin{equation}
 \phi = L i,
\end{equation}

donde $L$ es una constante que depende de los factores geométricos y de entorno llamada *inductancia*.

Los cambios de flujo electromagnético originan potenciales eléctricos relacionados por la **Ley de Faraday**

\begin{equation}
 u_{L} = - \frac{\mathrm{d}\phi}{\mathrm{d}t},
\end{equation}

donde $u_{L}$ denota el voltaje en las terminales de la inductancia a razón del cambio de flujo. Por consiguiente, la Ley de Faraday se puede expresar como sigue

\begin{equation}
 u_{L} = - L \frac{\mathrm{d}\phi}{\mathrm{d}t}.
\end{equation}

En elementos resistivos el voltaje $u_{R}$ entre el componente y la corriente $i$ que circula por él obedecen a la **Ley de Ohm** dada como siguiente

\begin{equation}
 u_{R} = Ri,
\end{equation}

donde $R$ es una constante que depende del componente denominado resistencia.

El voltaje $u_{C}$ entre las terminales de una capacitancia y la carga $q$ siguen la siguiente relación

\begin{equation}
 u_{C} = \frac{q}{C} \equiv \frac{1}{C} \int i \mathrm{d}t,
\end{equation}

donde $C$ es una constante que depende de la geometría y el entorno denominada capacitancia. Si consideramos que la corriente se define como una variación temporal de carga

\begin{equation}
 i = \frac{\mathrm{d}q}{\mathrm{d}t}, \quad q = \int i \mathrm{d}t,
\end{equation}

entonces $u_{C}$ se expresa en los siguientes términos

\begin{equation}
 u_{C} = \frac{1}{C} \int i\mathrm{d}t.
\end{equation}

### Leyes de Kirchoff
Establecen dos relaciones fundamentales en el análisis de circuitos eléctricos.

### Ley de malla
La suma de los voltajes en todo lazo cerrado de un circuito eléctrico es nula.

### Ley de nodos
La suma de las corrientes en todo punto de un circuito eléctrico es nula.

Las leyes de Kirchoff permiten establecer las ecuaciones dinámicas que siguen las corrientes y voltajes en los
circuitos eléctricos.

#### Ejemplo 1

Considere el circuito considerado en la siguiente figura

Por la *Ley de Malla*, obtenemos

\begin{equation}
 u_{R} + u_{C} = V_{in},
\end{equation}

por las relaciones anteriores, tenemos

\begin{equation}
 Ri + \frac{1}{C} \int i \mathrm{d}t = V_{in},
\end{equation}

expresado en términos de la carga $q$, tenemos la siguiente expresión

\begin{equation}
 R \dot{q} + \frac{1}{C}q = V_{in},
\end{equation}

Por la *Ley de Nodos*, tenemos

\begin{equation}
 i_{R} + i_{C} = i,
\end{equation}

dado que el voltaje entre los componentes eléctricos es el mismo y lo denotamos por $u$, tenemos

\begin{equation}
 \frac{u}{R} + C \dot{u} = i.
\end{equation}

Considerando el modelo obtenido por la *Ley de Malla* y considerando $V$ como la carga en el capacitor dividida
por la capacitancia $V = q/C$, sustituimos

\begin{equation}
 \begin{aligned}
  R\dot{V}C + \frac{1}{C}VC &= V_{in}, \\
  R\dot{V}C + V &= V_{in}, \\
  \dot{V} + \frac{1}{RC}V &= \frac{1}{RC} V_{in}.
 \end{aligned}
\end{equation}

La función de transferencia de este sistema está dada por la siguiente ecuación

\begin{equation}
 \frac{V(s)}{U(s)} = \frac{1}{RCs + 1} \equiv \frac{1}{\tau s + 1},
\end{equation}

donde $\tau = RC$.

La solución de una ecuación diferencial lineal

\begin{equation}
 \dot{y}(x) + p(x)y(x) = q(x),
\end{equation}

está dada por una familia uniparamétrica de la forma

\begin{equation}
 y(x) = \frac{1}{\mu(x)}\left[ \int_{0}^{x} \mu(t)q(t)\mathrm{d}t + c \right],
\end{equation}

donde

\begin{equation}
 \mu(x) = \exp\left[ \int_{0}^{x} p(t)\mathrm{d}t \right].
\end{equation}

Considere la ec. dif. del circuito $R$C en términos del voltaje que hay a través del capacitor

\begin{equation}
 \dot{V}(t) + \frac{1}{RC}V(t) = \frac{1}{RC}V_{in}(t).
\end{equation}

Considerando $p(t) = \frac{1}{RC}$ y $q(t) = \frac{1}{RC} V_{in}(t)$, tenemos

\begin{equation}
 y(t) = \frac{1}{\exp\left[ \int_{0}^{t}\frac{1}{RC}\mathrm{d}x \right]} \left[ \int^{t} \exp\left[ \int^{t} \frac{1}{RC}\mathrm{d}x \right] \frac{1}{RC}V_{in}\mathrm{d}x + c_{1} \right]
\end{equation}

resolviendo las integrales

\begin{equation}
 \begin{aligned}
  y(t) &= e^{-\frac{t}{RC}}\left(V_{in}e^{\frac{t}{RC}} + c_{1} \right),
  y(t) &= V_{in} + e^{-\frac{t}{RC}}c_{1}.
 \end{aligned}
\end{equation}

Considerando c.i. iguales a cero $c_{1} = 0$. Por consiguiente

\begin{equation}
 y(t) = V_{in}\left( 1 - e^{-\frac{t}{RC}} \right).
\end{equation}
